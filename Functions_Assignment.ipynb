# what is difference between a function and a method in python.

- A function is defined using the def keyword and is not associated with any object. It can be called independently.

Example:

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))

greet is a function that takes a parameter and returns a greeting.


- A method is a function that is associated with an object, typically defined inside a class. It is called on an instance of the class and usually takes self as its first parameter.

Example:

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Alice")
print(p.greet())


greet here is a method because it is part of the Person class and is called with an object (p.greet()).

# What are iterators in Python and how do they differ from iterables?

- In Python, iterables and iterators are key concepts used when working with loops (especially for loops). They are related but not the same.

    - Iterable:
        - An iterable is any Python object that can return an iterator.It must implement the __iter__() method.

        Example:
my_list = [1, 2, 3]
for item in my_list:
        
    print(item)
* Here, my_list is an iterable because it can be looped over.

    - Iterators:
        - An iterator is an object that represents a stream of data; it returns items one at a time using the __next__() method, and keeps track of its current position.

It must implement both:
  - __iter__() → returns the iterator object itself.
  - __next__() → returns the next item, or raises StopIteration when done.

      Example:

my_list = [1, 2, 3]
iterator = iter(my_list)  # Convert iterable to iterator

print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3

* next(iterator) would now raise StopIteration


#  Explain the concept of generators in Python and how they are defined.

- A generator is a special type of iterator in Python used to generate values on the fly instead of storing them in memory all at once.

Generators are:
    - Memory-efficient
    - Lazy (produce items only when needed)
    - Defined using def and the yield keyword (instead of return)

- How to define a generator:

Example: using a function with yield:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
* This function returns a generator object when called.

# What are the advantages of using generators over regular functions?

-  Advantages of Using Generators Over Regular Functions in Python: Generators offer several powerful advantages, especially in situations involving large datasets, streaming data, or performance optimization.

    - Memory Efficiency: Regular functions often return full lists, which consume a lot of memory. Generators yield one item at a time, which saves memory.

Example:

  - Regular Function:

def get_numbers_list(n):
    return [i for i in range(n)]

numbers = get_numbers_list(1000000)  # Loads entire list in memory

  - Generator Function:

def get_numbers_list(n):
    return [i for i in range(n)]

numbers = get_numbers_list(1000000)  # Loads entire list in memory

* Advantage: Generator avoids storing all 1 million numbers in memory.

    - Faster Startup time: Regular functions compute all results before returning. Generators start immediately and yield values one at a time.

Example:

def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(3):
    print(i)

* No need to wait for all values to be ready — they stream out as needed.

    - Infinite sequences: Generators can represent infinite series, which regular functions cannot without crashing your memory.

Example:

def infinite_counter():
    n = 1
    while True:
        yield n
        n += 1

counter = infinite_counter()
for _ in range(5):
    print(next(counter))

* This generator never ends, but we control how much we consume.


# What is a lambda function in Python and when is it typically used?

- A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can have any number of arguments, but only one expression.

- Lambda functions are typically used when:
    - You need a short function for a short period.
    - You want to avoid defining a full function using def.
    - Used as arguments to functions like map(), filter(), sorted(), etc.

Example:

    - With map()

numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # Output: [1, 4, 9, 16]

    - With filter()

numbers = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)  # Output: [2, 4]


    - With sorted() and key argument:

people = [("Alice", 25), ("Bob", 20), ("Eve", 30)]
sorted_people = sorted(people, key=lambda x: x[1])
print(sorted_people)  # Sorted by age


#  Explain the purpose and usage of the `map()` function in Python.

- The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new map object (an iterator).

- Purpose of map():
    - To transform each item in an iterable using a given function.
    - Useful for element-wise operations without writing loops.

Examples:

    - Using map() with a Built-in Function:

numbers = ["1", "2", "3"]
result = map(int, numbers)
print(list(result))  # Output: [1, 2, 3]

 * Converts each string to an integer

    - Using map() with a Lambda Function:

numbers = [1, 2, 3, 4]
squares = map(lambda x: x ** 2, numbers)
print(list(squares))  # Output: [1, 4, 9, 16]

  * Applies x ** 2 to each element.

  

# What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- map() – Transform Each Item
    - Applies a function to each item in an iterable.
    - Returns a new map object (iterator).

Example:

numbers = [1, 2, 3, 4]
squares = map(lambda x: x * x, numbers)
print(list(squares))  # Output: [1, 4, 9, 16]

  (Use when you want to transform or modify items.)



- filter() – Select Items That Meet a Condition
    - Applies a function to each item.
    - Returns only those items for which the function returns True.

Example:

numbers = [1, 2, 3, 4, 5]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4]

 (Use when you want to filter items based on a condition.)


- reduce() – Reduce to a Single Value
    - Applies a function to the first two items, then to the result and next item, and so on.
    - Returns a single value.
    - Must be imported from functools.

Example:

from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24


(Use when you want to combine items into one value, like a sum or product.)

# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

my_list = [1, 4, 10, 15, 20, 45, 8]
result = sum_even_numbers(my_list)
print("Sum of even numbers:", result)
```

#  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def square_numbers(numbers):
    squared_list = []
    for num in numbers:
        squared_list.append(num ** 2)
    return squared_list


input_list = [1, 2, 3, 4, 5, 6, 7]
result = square_numbers(input_list)
print("Squared numbers:", result)
```

# Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def is_prime(n):
    if n < 2 or n > 200:
        return False  # Not in valid range or not prime
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 35):
    print(f"{num} is prime? {is_prime(num)}")

```

# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # Stop when max_terms is reached
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a


fib = Fibonacci(10)  # Generate first 10 Fibonacci numbers

for num in fib:
    print(num, end=' ')
```

# Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Get powers of 2 up to 2^5
for power in powers_of_2(5):
    print(power, end=' ')
```

# Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

# dummy file for demonstration
with open('example.txt', 'w') as f:
    f.write('Line 1: Hello, world!\n')
    f.write('Line 2: This is a sample text file.\n')
    f.write('Line 3: Reading lines one by one.\n')


file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line, end='')
```

#  Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.

```python
# fromula to convert celsius to fahrenheit = F=(C*9/5)+32

# List of temperatures in Celsius
celsius = [0, 10, 20, 35, 50, 70, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```

#  Create a Python program that uses 'filter()' to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello World, My Name is Tushar Sharma"
result = remove_vowels(text)
print("Original string:", text)
print("Without vowels: ", result)
```
